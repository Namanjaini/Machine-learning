```python

import pandas as pd

df = pd.read_csv(r"C:\Users\user\Documents\Machine learning\machine-learning-ex1\ex1\ex1data2.txt",names = ['House Size(Square feet)',"#Bedrooms",'Price'])
df
```

```python
x1 = df['House Size(Square feet)']
x2 = df['#Bedrooms']
y = df['Price']


print(x1)
print(x2)
print(y)

import numpy as np
lst = [1 for i in range(len(x1))]

#creating the matrix X by adding column of ones
X = np.array([lst,x1,x2])
X = X.transpose()    ####this transpose is used for making the matrix 
print(X)

Y = np.array(y)
Y = Y.transpose()
Y = Y.reshape(len(y),1)
print(Y)
```

```python
XT=X.transpose()

print(XT)
print (X)

a=XT.dot(X)
print(a)
inverse_a = np.linalg.inv(a)   ##inverse of XT*X
print(inverse_a)

b=XT.dot(Y)
print (b)

c=inverse_a.dot(b)
theta=c
theta
```
