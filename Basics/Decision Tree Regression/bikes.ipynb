```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
bikes = pd.read_csv('bikes.csv')
bikes.head()
```

```python
plt.figure(figsize=(8,6))
plt.plot(bikes['temperature'], bikes['count'], 'r.')
plt.xlabel('temperature')
plt.ylabel('bikes')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(max_depth=2)
regressor.fit(np.array([bikes['temperature']]).T, bikes['count'])
```

```python
print(regressor.predict(2))
print(regressor.predict(6))
print(regressor.predict(10))
print(regressor.predict(13))
print(regressor.predict(15))
print(regressor.predict(20))
```

```python
x_grid = np.array([np.linspace(-5, 40, 100)]).T
plt.figure(figsize=(8,6))
plt.plot(bikes['temperature'], bikes['count'], 'r.', label='observation')
plt.plot(x_grid, regressor.predict(x_grid), linewidth=4, alpha=.7, label='prediction')
plt.xlabel('temperature')
plt.ylabel('bikes')
plt.legend()
plt.show()
```
